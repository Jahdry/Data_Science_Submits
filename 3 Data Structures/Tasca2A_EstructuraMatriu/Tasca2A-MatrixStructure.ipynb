Crea un np.array d'una dimensió, que inclogui l'almenys 8 nombres sencers, data type int64. Mostra la dimensió i la forma de la matriu. .


### Exercici 3

Crea una matriu bidimensional amb una forma de 5 x 5. Extreu el valor màxim de la matriu, i els valors màxims de cadascun dels seus eixos.

```python
#Generem la matru: 
arr_2d = np.random.randint(100, size=(5,5))

#Definim funció per calcular els maxims i minims de lineas i columnes: 
def line_max(arr_2d): 
    for y in range(len(arr_2d)):
        line_max = arr_2d[y].max()
        print(f'line {y+1} max value is: {line_max}')
    print('\n')

def column_max(arr_2d): 
    files, columns = arr_2d.shape
    for y in range(files):
        colum = []
        for x in range(columns): 
            colum.append(arr_2d[x, y])
            colum_arr = np.array(colum)
        #print(colum_arr)
        print(f'col. {y+1} max value is: {colum_arr.max()}')
    print('\n')

#Calculem el maxim de la matriu: 
matrix_max = arr_2d.max()

#Outputs
print('array:\n\n', arr_2d)
print('\n number of dimensions :', arr_2d.ndim,', shape:', arr_2d.shape )
print()
print(f'matrix max value is: {matrix_max}\n')
line_max(arr_2d)
column_max(arr_2d)
```

### Exercici 5

Utilitza la Indexació per extreure els valors d'una columna i una fila de la matriu. I suma els seus valors.

```python
# Definim la matriu
arr1 = np.random.randint(18, size=(4, 3), dtype='int64') 

print(arr1)
#Extraiem valors:
fila = arr1[3] #tots els index [3][x]
columna = arr1[0:,2] #tots els index [0-5][2]

# Output + suma dels valors de cada extraacció
print()
print('fila:', fila, 'suma:', fila.sum())
print('columna:', columna, 'suma:', columna.sum())
```

### Exercici 6

Mask la matriu anterior, realitzeu un càlcul booleà vectoritzat, agafant cada element i comprovant si es divideix uniformement per quatre.

Això retorna una matriu de mask de la mateixa forma amb els resultats elementals del càlcul.

```python
mask = arr1 % 4 == 0 #Generem la mascara indicant les condicions
print('Matriu de mascara:\n\n',mask)
```

### Exercici 7

A continuació, utilitzeu aquesta màscara per indexar a la matriu de números original. Això fa que la matriu perdi la seva forma original, reduint-la a una dimensió, però encara obteniu les dades que esteu cercant.

```python
masked_arr = arr1[mask] #indexem la matriu de mascara a la matriu original per veure els valors que si que compleixen la condició
print(masked_arr)
```

Mostreu-me a veure que passa quan eliminem el canal G Verd o B Blau.

Mostreu-me a veure què passa quan eliminem el canal G Verd o B Blau. Hauries d'utilitzar la indexació per seleccionar el canal que voleu anul·lar.

Utilitzar el mètode, mpimg.imsave () de la llibreria importada, per guardar les imatges modificades i que haureu de pujar al vostre repositori a github.

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt


spidey_img = mpimg.imread('Spidey.jpg') #importem imatge
print(spidey_img) #array de la imatge amb els canals RGB
```

```python
imgplot = plt.imshow(spidey_img)
```

```python
#Eliminem el canal verd: 

spidey_noblue= spidey_img.copy()
spidey_noblue[0:, 0:,2]=0
print(spidey_noblue)
```

```python
imgplot = plt.imshow(spidey_noblue)
mpimg.imsave('spidey_noblue.jpg', spidey_noblue)
```

```python
spidey_nogreen= spidey_img.copy()
spidey_nogreen[0:, 0:,1]=0
imgplot = plt.imshow(spidey_nogreen)
mpimg.imsave('spidey_nogreen.jpg', spidey_nogreen)
```
