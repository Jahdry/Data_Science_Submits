## **BEAUTIFUL SOUP**

```python
response = requests.get("https://www.bolsamadrid.es/esp/aspx/Mercados/Precios.aspx?indice=ESI100000000")
soup = BeautifulSoup(response.text, "html.parser")
# print(soup.text)
# print(soup.prettify)
```

```python
ibex35 = soup.find(class_="TblPort", id="ctl00_Contenido_tblAcciones")
ibex35.text
```

```python
table_rows = ibex35.find_all("tr")

lines = []
items =[]
for row in table_rows: 
    attributes = row.find_all(["th","td"])
    items = []
    for att in attributes:
        items.append(att.text.strip())
    lines.append(items)
print (lines)
```

```python
stocks_dataframe = pd.DataFrame(lines[1:], columns=lines[0])
stocks_dataframe.to_csv('../db/stocks_IBEX35_09_20_2021_cierre sesion.csv')
stocks_dataframe
```

```python
class OfertasBadalona(scrapy.Item):
    # define the fields for your item here like: 
    # name = scrapy.Field()

    #info de la oferta
    name    = scrapy.Field()
    desc    = scrapy.Field()
    period  = scrapy.Field()
    regim    = scrapy.Field()
    specs    = scrapy.Field()
    
class OfertsSpider(Spider):
    name = 'ofertas'
    item_count = 0
    allowed_domain = ['http://badalona.cat']
    start_urls = ['http://badalona.cat/portalWeb/badalona.portal?_nfpb=true&_pageLabel=contingut_estatic&dCollectionID=4362#wlp_contingut_estatic']



    def parse(self, response):
        ofertes = response.xpath('//h2[@class="h2_2n "]/a/@href').getall()

        for ofertes_item in ofertes:
            oferta_item = OfertasBadalona()
            #info de la oferta
            oferta_item['name']     = response.xpath('/html/body/div[5]/div/div[1]/div/h1/text()').extract()
            oferta_item['desc']     = response.xpath('/html/body/div[5]/div/div[1]/div/div/div[1]/div[1]/p[1]/text()').extract()
            oferta_item['period']   = response.xpath('/html/body/div[5]/div/div[1]/div/div/div[1]/div[1]/p[2]/text()').extract()
            oferta_item['regim']    = response.xpath('/html/body/div[5]/div/div[1]/div/div/div[1]/div[1]/p[3]/text()').extract()
            oferta_item['specs']    = response.xpath('/html/body/div[5]/div/div[1]/div/div/div[1]/div[1]/p[4]/text()').extract()
            #self.item_count += 1
            #if self.item_count > 10:
                # raise CloseSpider('items exceeded')
            yield oferta_item
            filename = 'ofertes.csv'
            with open(filename, 'wb') as f:
                for key in oferta_item.keys():
                    f.write(f'{key}, {oferta_item[key]}')
```

```python
process = CrawlerProcess({
    'FEED_FORMAT': 'csv', 
    'FEED_EXPORT_FIELDS': ['Title', 'Description', 'Period', 'Regime', 'Specifications'],
    'FEED_URI': 'file://./Users/Jahdry/Documents/GitHub/Data_Science_Submits/Sprint_16/export.csv',
})
process.crawl(OfertsSpider)
process.start()
```
