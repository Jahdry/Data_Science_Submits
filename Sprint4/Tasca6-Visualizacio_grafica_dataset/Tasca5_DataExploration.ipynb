#### Fes un informe complet del data set:.

```python
# Comprovem les dades faltants que hi ha a cada columna
flights_df_filtered.isnull().sum()
```

```python
#Netegem les files que contenten alguna columna sense dades i creem un nou view 
flights_df_clean = flights_df_filtered.dropna(inplace=False)
#Comprovem que el nou view no te cap data vuida
flights_df_clean.isnull().sum()
```

* Resumeix estadísticament les columnes d’interès

```python
pd.set_option('display.float_format', lambda x: '%.2f' % x) #Configurem el display per als numeros decimals amb 2 possicions
flights_df_clean.describe()
```

* Crea columnes noves (velocitat mitjana del vol, si ha arribat tard o no...)

```python
# Creem les series de dades per a cada una de les columnes que volem agefir
mean_vel = flights_df_clean['Distance']/ (flights_df_clean['AirTime']/60) #Velocitat mitja
del_arr = flights_df_clean['ArrDelay'] > 0 # Si el vol ha arribat tard (bool)

# del_total = flights_df_clean['ArrDelay'] + flights_df_clean['DepDelay'] #Enderreriment total. Havia creat aquesta columna per calcular els excercicis seguents pero després he decidit que es mes acurat fer servir el 'ArrDelay' ja que es el retard realment important.
```

```python
flights_df_clean = flights_df_filtered.dropna(inplace=False) #Torno a possar aquesta funció aqui perque em generi l'array de nou cada cop que faig correr aquesta cel·la i evitar l'error de que les col·lumnes ja existeien

#Insertem les series com a noves columnes al final de DF.
flights_df_clean.insert(15, "Vel.Mean(km/h)", mean_vel)
flights_df_clean.insert(16, "DelayAtArribal", del_arr)
flights_df_clean
```

* Taula de les aerolínies amb més endarreriments acumulats

```python
is_delayed = flights_df_clean['DelayAtArribal'] # Creem un array de boleans per al vols que han arribat tard
delayed_flights = flights_df_clean[is_delayed] # Apliquem l'array de voleans per tenir un nou array on nomes es mostrin els vols que han arribat tard
delayed_flights[:5]
```

```python
delay_cum = delayed_flights[['UniqueCarrier','ArrDelay']]
delay_cum_grouped = delay_cum.groupby('UniqueCarrier').aggregate([sum])
delay_cum_grouped_sorted = delay_cum_grouped.sort_values(by=('ArrDelay', 'sum'), ascending=False)
print(f'Les 5 aerolinies amb mes enderreriments acumulats (mins) son:\n{delay_cum_grouped_sorted[:5]}') 

### Un altre manera de fer-ho: fem l¡agregate directament de tot el dataframe i despres filtrem per la columna que volem
### cumulate_delay = flights_df_clean.groupby('UniqueCarrier').aggregate(sum)
### cumulate_delay['ArrDelay'].sort_values()
```

* Taula de les Aerolinies amb mes enderriments en %

```python
flights_by_uc = flights_df_clean['UniqueCarrier'].value_counts() #Creem una serie amb els counts del vols ordenats per 'Unique Carrier'
del_flights_by_uc = delayed_flights['UniqueCarrier'].value_counts() # Serie amb els counts dels vols que han arribat tard ordenats per 'Unique Carrier'
del_flights_percent = (del_flights_by_uc / flights_by_uc * 100).sort_values(ascending=False) # dividim els vols retrasats entre els vols totals per tenir el % de vols retrasats per cada companyia

# Representem graficament:
print(f'Les tres compañies amb un major percentatge de vols enderrerits son: \n{del_flights_percent[:3]}') 
del_flights_percent.plot(kind='bar') 
```

* Quins són els vols més llargs?

    En total: 

```python
#fem un view del df on es mostrin les files ordenades pel temps que ha estat l'avió enlairat i mostrem els 5 vols amb el temps mes llarg
longest_flights = flights_df_clean[['UniqueCarrier','FlightNum','TailNum','AirTime']].sort_values(by='AirTime', ascending=False)
print(f'els 5 vols mes llargs son:\n{longest_flights[:5]}')
```

El vol mes llarg de cada companyia: 

```python
# Agrupem els vols per compania
longest_flights_grouped = longest_flights.groupby('UniqueCarrier')

for name, group in longest_flights_grouped: # en cada grup iterem perque mostri el contingut del grup
    print ('Unique Carrier:',name)
    print (group.sort_values(by='AirTime', ascending=False)[:1]) # ordenem i filtrem perque només mostri el vol mes llarg de cada grup
    print()
```

* Quins son els vols més endarrerits?

```python
#Entenc que el Dep Delay pot ser compensat durant el vol així que el delay que compta finalment el d'arribada 'ArrDelay'

#fem un nou view del df ordenat per 'ArrDelay' veure els vols que han arribat mes tard
most_delayed_flights = delayed_flights[['UniqueCarrier','FlightNum','TailNum','ArrDelay']].sort_values(by='ArrDelay', ascending=False)[:5]
print(f'els 5 vols mes endarrerits son:\n{most_delayed_flights}')
```
